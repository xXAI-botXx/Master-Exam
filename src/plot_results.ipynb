```python
for idx, cur_data in enumerate(data):
    avg = np.array(cur_data).mean()
    experiment_data[idx]["loss"] = avg

experiment_data
```

### Visual Analysis

```python
df = pd.DataFrame(experiment_data)

corr = df.drop(columns=['ID']).corr()

score_corr = corr[['loss']].drop('loss')

plt.figure(figsize=(8, 5))
sns.heatmap(score_corr, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation of the weights with the Loss')
plt.show()
```

```python
score_corr['abs_corr'] = score_corr['loss'].abs()
score_corr = score_corr.sort_values('abs_corr', ascending=False)

plt.figure(figsize=(8, 5))
sns.barplot(x=score_corr.index, y=score_corr['loss'], palette='coolwarm')
plt.xticks(rotation=45)
plt.ylabel('Correlation with Loss')
plt.xlabel('Loss Weights')
plt.title('Influence from the loss-weights towards the loss')
plt.axhline(0, color='gray', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
for col in df.columns:
    if col.startswith('weight_'):
        sns.scatterplot(data=df, x=col, y='loss')
        plt.title(f'{col} vs. Score')
        plt.show()
```
